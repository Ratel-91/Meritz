```python
import os
os.getcwd()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("KTB_data.xlsx", sheet_name= 'data set')
tenor = pd.read_excel("KTB_data.xlsx", sheet_name= 'tenor', header = None)
tenor.columns = ['item', 'tenor']
dates = data['Date']
items = data.columns
```

```python
carry_pa = pd.DataFrame(columns= items)
roll_3m = pd.DataFrame(columns= items)
cr_3m = pd.DataFrame(columns= items)
cr_3m_dv01 = pd.DataFrame(columns= items)

carry_pa['Date'] = dates
roll_3m['Date'] = dates
```

```python
# Calculate Carry p.a
for i in data.index:
    carry_pa.iloc[i, 2:] = data.iloc[i, 2:] - data.iloc[i, 1]
```

```python
# Calculate 3m Rolldown
for i in range(2, 11):
    roll_3m.iloc[:, i] = (data.iloc[:, i] - data.iloc[:, i - 1]) / (tenor.iloc[i - 1, 1] - 
                                               tenor.iloc[i - 2, 1]) * 3/12 * (tenor.iloc[i - 1, 1] -3/12)
    
for j in range(13, 21):
    roll_3m.iloc[:, j] = (data.iloc[:, j] - data.iloc[:, j - 1]) / (tenor.iloc[j - 1, 1] - 
                                               tenor.iloc[j - 2, 1]) * 3/12 * (tenor.iloc[j - 1, 1] -3/12) 
```

```python
# 3m Gross Carry and Rolldown, date끼리 더할 수 없어서 여기서 Date가 컬럼에서 빠짐 
cr_3m = roll_3m.iloc[:,2:] + carry_pa.iloc[:,2:]/4
```

```python
# 3m Gross Carry and Rolldown per DV01
cr_3m_dv01 = roll_3m.iloc[:,2:] + carry_pa.iloc[:,2:]/4
for i in range(0, 10):
    cr_3m_dv01.iloc[:, i] = cr_3m.iloc[:, i] / tenor.iloc[i + 1, 1]
                                               
for j in range(11, 20):
    cr_3m_dv01.iloc[:, i] = cr_3m.iloc[:, i] / tenor.iloc[i + 1, 1]
```

```python
cr_3m_dv01.index = dates
cr_3m_dv01 = cr_3m_dv01.fillna(0)
```
